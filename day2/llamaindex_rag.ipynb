```python
import os
from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')
```

```python
os.environ['SAMBANOVA_API_KEY'] = ""
```

```python
import qdrant_client
from llama_index.core import Settings
from llama_index.core import SimpleDirectoryReader
from llama_index.embeddings.fastembed import FastEmbedEmbedding
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.core import StorageContext,VectorStoreIndex
from llama_index.vector_stores.qdrant import QdrantVectorStore
```

```python
from llama_index.llms.sambanovasystems import SambaNovaCloud

llm2 = SambaNovaCloud(
    model="DeepSeek-R1",
    context_window=100000,
    max_tokens=1024,
    temperature=0.1,
)
```

```python
documents = SimpleDirectoryReader("./data").load_data()
```

```python
QDRANT_API_KEY = userdata.get("qdrant_key")
QDRANT_URL = userdata.get("qdrant_url")
```

```python
client = qdrant_client.QdrantClient(
    url=QDRANT_URL,
    api_key=QDRANT_API_KEY,
    prefer_grpc = True
)
```

```python
Settings.llm = llm2
```

```python
Settings.embed_model = FastEmbedEmbedding(model_name="BAAI/bge-base-en-v1.5")
```

```python
vector_store = QdrantVectorStore(client=client, collection_name="bootcamp")
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context,
)
```

```python
query_engine = index.as_query_engine()
response = query_engine.query("Where is Tarun Jain currently working?")
```
